```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Ellipse
import math
```

```python
path = "~/Developer/probp.finalproject.darkMatter/datasets"
halo_path = "Training_halos.csv"
halo = (pd.read_csv(path+"/"+halo_path).values)
```

```python

#plt.scatter()
sky = 1
for sky in np.arange(1,301,1):
    plt.clf()
    path = "~/Developer/probp.finalproject.darkMatter/datasets"
    sky_path = "Train_Skies/Training_Sky"+str(sky)+".csv"
    halo_path = "Training_halos.csv"
    a = (pd.read_csv(path+"/"+sky_path).values)
    halos = np.array(halo[:,1:])

    fig = plt.figure(0)
    ax = fig.add_subplot(111,aspect='equal')
    ax.set_xlim(0, 4200)
    ax.set_ylim(0, 4200)

    xy = np.array(a[:,1:3])
    #print(xy)
    e1,e2 = np.array(a[:,3:4],dtype=np.float32),np.array(a[:,4:5],dtype=np.float32)
    angle = np.arctan2(e2,e1)/2.
    angle_pi = angle/(2.* math.pi)*360.
    sq = np.sqrt(e1**2+e2**2)
    q = ((1.+sq)/(1-sq))**2
    a1 = np.sqrt(q/(1+q))
    b1 = np.sqrt(1./(1+q))
    #print (b1)
    cons = 100
    #plt.clf()
    for i in range(xy.shape[0]):
        if a1[i]/b1[i] >2:
            e = Ellipse(xy=xy[i], width=a1[i]*cons, height=b1[i]*cons, angle=angle_pi[i])
            e.set_clip_box(ax.bbox)
            e.set_alpha(1)
            ax.add_artist(e)
            e.set_facecolor((0,0,0))
            for j in range(halos[sky-1][0]):
                    e = Ellipse(xy=halos[sky-1][j*2+1:j*2+3], width=300, height=300, angle=0)
                    e.set_clip_box(ax.bbox)
                    e.set_alpha(0.5)
                    ax.add_artist(e)
                    e.set_facecolor((1,0,0))
    plt.savefig("filtered_sky"+str(sky)+".pdf")
   
#     plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse

NUM = 250

ells = [Ellipse(xy=rnd.rand(2)*10, width=rnd.rand(), height=rnd.rand(), angle=rnd.rand()*360)
        for i in range(NUM)]

fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

plt.show()
```
